```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("/content/clean_frality_yield.csv")
```

```python
#printing the data
data
```

```python
#It plots the each unique value in the frailty column
sns.countplot(x='Frailty', data=data)
plt.title('Frailty Distribution')
plt.show()
```

```python
#It plots the scatter plot of weight versus height
plt.scatter(data['Height(Inches)'],data['Weight(Pounds)'], alpha=0.5)
plt.xlabel('Height(Inches)')
plt.ylabel('Weight(Pounds)')
plt.title('Weight vs. Height')
plt.show()
```

```python
ax = sns.heatmap(data.corr(), cmap="YlGnBu", linewidths = 0.30, annot = True)
fig = ax.get_figure()
```
